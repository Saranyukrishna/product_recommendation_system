```python
def split_and_comma(s):
    words = re.findall(r'[A-Z]?[a-z]+|[0-9]+', s) 
    return ', '.join(words)
df['product_name'] = df['product_name'].apply(split_and_comma)
print(df[['product_name']])
```

```python
df['product_category_tree']
```

```python
df['product_category_tree'] = df['product_category_tree'].apply(lambda x: x.split(' >> '))
print(df['product_category_tree'])
```

```python
def clean(s):
    if isinstance(s,list):
        s=''.join(s)
    s = re.sub(r'[\[\],]', '', s)
    s=s.replace('&', '')
    s = re.sub(r'[^a-zA-Z0-9\s]', '', s)
    words=s.split()
    s=','.join(words)
    return s.strip()
df['product_category_tree']=df['product_category_tree'].apply(lambda x: clean(x))
print(df['product_category_tree'])
```

```python
import nltk
nltk.download('stopwords')
```

```python
import nltk 
from nltk.corpus import stopwords
st=set(stopwords.words('english'))
def remove_stopwords(text):
    if isinstance(text,str):

        words=text.split()
        f=[word for word in words if word.lower() not in st]
        return ''.join(f)
    return text    
```

```python
x=['product_category_tree','product_name','brand']
for i in x:
    df[i]=df[i].apply(remove_stopwords)
    df = df.applymap(lambda x: x.replace("'", "") if isinstance(x, str) else x)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer=TfidfVectorizer(stop_words='english')
tfidf_matrix=tfidf_vectorizer.fit_transform(df['product_category_tree'])
cosine_similarities=cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
df['product_name'][0]
```

```python
x='Alisha,Solid,Women,s,Cycling,Shorts'
item_index=df[df['product_name']==x].index[0]
```

```python
similar_items=list(enumerate(cosine_similarities[item_index]))
```

```python
similar_items=sorted(similar_items,key=lambda x:x[1],reverse=True)
top_similar_items=similar_items[1:10]
reccomended=[x[0] for x in top_similar_items]
```
